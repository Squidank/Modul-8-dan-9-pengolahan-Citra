## Sobel

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca citra menggunakan OpenCV
img = cv2.imread('conan.jpg')

# Menggunakan operator Sobel untuk deteksi tepi pada sumbu x dan sumbu y
sobelx = cv2.Sobel(img, cv2.CV_8U, 1, 0, ksize=5)
sobely = cv2.Sobel(img, cv2.CV_8U, 0, 1, ksize=5)

# Menjumlahkan hasil deteksi tepi pada sumbu x dan sumbu y
sobel = sobelx + sobely

# Membuat subplot dengan ukuran 4x2 untuk menampilkan citra dan histogramnya
fig, axes = plt.subplots(4, 2, figsize=(20, 20))
ax = axes.ravel()

# Menampilkan citra input dan histogramnya
ax[0].imshow(img, cmap='gray')
ax[0].set_title("input")
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title("histogram")

# Menampilkan citra hasil deteksi tepi pada sumbu x dan histogramnya
ax[2].imshow(sobelx, cmap='gray')
ax[2].set_title("output")
ax[3].hist(img.ravel(), bins=256)
ax[3].set_title("histogram")

# Menampilkan citra hasil deteksi tepi pada sumbu y dan histogramnya
ax[4].imshow(sobely, cmap='gray')
ax[4].set_title("output")
ax[5].hist(img.ravel(), bins=256)
ax[5].set_title("histogram")

# Menampilkan citra hasil penjumlahan deteksi tepi sumbu x dan sumbu y (total deteksi tepi) dan histogramnya
ax[6].imshow(sobel, cmap='gray')
ax[6].set_title("output")
ax[7].hist(img.ravel(), bins=256)
ax[7].set_title("histogram")

fig.tight_layout()
```

## Prewit

```python
# Membuat kernel Prewitt untuk deteksi tepi sumbu x dan sumbu y
kernelx = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
kernely = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])

# Menggunakan operator Sobel dengan kernel Prewitt untuk deteksi tepi sumbu x dan sumbu y
prewittx = cv2.Sobel(img, cv2.CV_8U, 1, 0, ksize=5)
prewitty = cv2.Sobel(img, cv2.CV_8U, 0, 1, ksize=5)

# Menjumlahkan hasil deteksi tepi sumbu x dan sumbu y
prewitt = prewittx + prewitty

# Membuat subplot dengan ukuran 4x2 untuk menampilkan citra dan histogramnya
fig, axes = plt.subplots(4, 2, figsize=(20, 20))
ax = axes.ravel()

# Menampilkan citra input dan histogramnya
ax[0].imshow(img, cmap='gray')
ax[0].set_title("input")
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title("histogram")

# Menampilkan citra hasil deteksi tepi sumbu x dan histogramnya
ax[2].imshow(prewittx, cmap='gray')
ax[2].set_title("output")
ax[3].hist(img.ravel(), bins=256)
ax[3].set_title("histogram")

# Menampilkan citra hasil deteksi tepi sumbu y dan histogramnya
ax[4].imshow(prewitty, cmap='gray')
ax[4].set_title("output")
ax[5].hist(img.ravel(), bins=256)
ax[5].set_title("histogram")

# Menampilkan citra hasil penjumlahan deteksi tepi sumbu x dan sumbu y (total deteksi tepi) dan histogramnya
ax[6].imshow(prewitt, cmap='gray')
ax[6].set_title("output")
ax[7].hist(img.ravel(), bins=256)
ax[7].set_title("histogram")

fig.tight_layout()
```

## Canny

```python
# Menggunakan operator Canny untuk deteksi tepi
canny = cv2.Canny(img, 100, 200)

# Membuat subplot dengan ukuran 2x2 untuk menampilkan citra dan histogramnya
fig, axes = plt.subplots(2, 2, figsize=(20, 20))
ax = axes.ravel()

# Menampilkan citra input dan histogramnya
ax[0].imshow(img, cmap='gray')
ax[0].set_title("input")
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title("histogram")

# Menampilkan citra hasil deteksi tepi menggunakan operator Canny dan histogramnya
ax[2].imshow(canny, cmap='gray')
ax[2].set_title("output")
ax[3].hist(img.ravel(), bins=256)
ax[3].set_title("histogram")

fig.tight_layout()
```

## Laplacian

```python
# Mengkonversi citra ke grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Menggunakan filter Gaussian untuk menghaluskan citra
img = cv2.GaussianBlur(gray, (3, 3), 0)

# Menggunakan operator Laplacian untuk deteksi tepi
laplacian = cv2.Laplacian(img, cv2.CV_64F)

# Menampilkan citra asli dan citra hasil deteksi tepi menggunakan operator Laplacian
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(img, cmap='gray')
plt.title('Laplacian Image')
```
