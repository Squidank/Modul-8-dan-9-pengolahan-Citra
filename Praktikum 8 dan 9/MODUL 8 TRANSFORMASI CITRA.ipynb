## 1. Transformasi Twirl 

```python
import matplotlib.pyplot as plt
#from skimage import data
from skimage.transform import swirl

image = plt.imread('vector.jpg') # membaca gambar dari direkt0ri
swirled = swirl(image, rotation=5, strength=6,radius=300) # fungsi swirl dengan settingan sesuai keinginan

fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(8, 3), sharex=True, sharey=True)
# Membuat figure dengan 1 baris dan 2 kolom subplot. Mengatur ukuran figure menjadi (8, 3) dengan mengaktifkan sharing pada sumbu x dan y.
# Menyimpan objek axes subplot pertama pada variabel ax0 dan subplot kedua pada variabel ax1.

ax0.imshow(image, cmap=plt.cm.gray)
ax0.axis('off')
# Menampilkan citra 'image' pada subplot pertama (ax0) dengan colormap gray.
# Menghilangkan tampilan sumbu pada subplot pertama (ax0).

ax1.imshow(swirled, cmap=plt.cm.gray)
ax1.axis('off')
# Menampilkan citra 'swirled' pada subplot kedua (ax1) dengan colormap gray.
# Menghilangkan tampilan sumbu pada subplot kedua (ax1).

plt.show()
# Menampilkan figure dengan semua subplot yang telah ditentukan.
```

## 2. Translation 

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread("King Salman.jpg")
# Membaca gambar "King Salman.jpg" menggunakan OpenCV dari direktori lokal

h, w = image.shape[0:2]
# Mengambil tinggi (h) dan lebar (w) citra dari properti shape.

half_height, half_width = h//4, w//8
# Menghitung setengah tinggi (half_height) dan setengah lebar (half_width) citra dengan pembagian bulat(dibulatkan).

transition_matrix = np.float32([[0.5, 0, half_width], [0, 0.7, half_height]])
# Membuat matriks transformasi 2x3 dengan koefisien [0.5(img menyusut setengah dari aslinya kesamping), 0, half_width] untuk translasi sumbu x
# dan [0,0.7 (img menyusut, 0.7 dari aslinya(vertikal)) , half_height] untuk translasi sumbu y menggunakan tipe data float32.

img_transition = cv.warpAffine(image, transition_matrix, (w, h))
# Menerapkan transformasi translasi pada citra menggunakan warpAffine dari OpenCV.
# fungsi affine memungkinakan untuk mengubah bentuk, rotasi, skala, dan translasi suatu gambar
# Citra "image" ditranslasi menggunakan matriks transformasi "transition_matrix" dengan ukuran (w, h)
# dan hasilnya disimpan dalam variabel "img_transition".

plt.imshow(cv.cvtColor(img_transition, cv.COLOR_BGR2RGB))
# Menampilkan citra "img_transition" dengan menggunakan colormap RGB.
# Perlu mengubah ruang warna BGR (yang digunakan oleh OpenCV) menjadi RGB (yang digunakan oleh matplotlib).

plt.title("Translation")
# Memberikan judul pada gambar.

plt.show()
# Menampilkan gambar.
```

```python
transition_matrix2 = np.float32([[0.5, 0.05, half_width], [0.1, 0.7, half_height]])
#  [0.5, 0.05(mengatur kemiringan horizontal), half_width] dan [0.1(mengatur kemiringan vertikal),0.7 , half_height]

img_transition2 = cv.warpAffine(image, transition_matrix2, (w, h))

# Membuat plot dengan 1 baris dan 2 kolom
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Menampilkan gambar pertama di subplot pertama (kolom pertama)
axes[0].imshow(cv.cvtColor(img_transition, cv.COLOR_BGR2RGB))
axes[0].set_title("Original Image")

# Menampilkan gambar kedua di subplot kedua (kolom kedua)
axes[1].imshow(cv.cvtColor(img_transition2, cv.COLOR_BGR2RGB))
axes[1].set_title("Translated Image")

# Menghilangkan label sumbu x dan y pada kedua subplot
#for ax in axes:
    #ax.axis("off")

# Menampilkan plot
plt.tight_layout()
plt.show()
```

## 3. Rotation

```python
image = cv.imread("King Salman.jpg")
# Membaca gambar "King Salman.jpg" menggunakan OpenCV dan menyimpannya dalam variabel "image".

h, w = image.shape[:2]
# Mengambil tinggi (h) dan lebar (w) citra dari properti shape.

rotation_matrix = cv.getRotationMatrix2D((w//4, h//2), -180, 0.5)
# Membuat matriks transformasi rotasi 2x3 dengan pusat rotasi pada koordinat (w//4, h//2),
# sudut rotasi sebesar -180 derajat (berlawanan arah jarum jam), dan skala sebesar 0.5.
# Fungsi getRotationMatrix2D dari OpenCV digunakan untuk menghasilkan matriks transformasi tersebut.

rotated_image = cv.warpAffine(image, rotation_matrix, (w, h))
# Menerapkan transformasi rotasi pada citra menggunakan warpAffine dari OpenCV.
# Citra "image" dirotasi menggunakan matriks transformasi "rotation_matrix" dengan ukuran (w, h)
# dan hasilnya disimpan dalam variabel "rotated_image".

plt.imshow(cv.cvtColor(rotated_image, cv.COLOR_BGR2RGB))
# Menampilkan citra "rotated_image" dengan menggunakan colormap RGB.
# Perlu mengubah ruang warna BGR (yang digunakan oleh OpenCV) menjadi RGB (yang digunakan oleh matplotlib).

plt.title("Rotation")
# Memberikan judul pada gambar.

plt.show()
# Menampilkan gambar.
```

```python
import mahotas as mh
# Mengimpor modul mahotas untuk pemrosesan citra.

from pylab import imshow, show
# Mengimpor fungsi imshow dan show dari modul pylab untuk menampilkan citra.

regions = np.zeros((8, 8), bool)
# Membuat matriks 8x8 yang berisi nilai False (0) menggunakan np.zeros dari NumPy.

regions[:3, :3] = 1
# Menetapkan nilai True (1) pada bagian kiri atas matriks regions dengan indeks baris 0-2 dan indeks kolom 0-2.

regions[6:, 6:] = 1
# Menetapkan nilai True (1) pada bagian kanan bawah matriks regions dengan indeks baris 6-7 dan indeks kolom 6-7.

labeled, nr_objects = mh.label(regions)
# Melabeli wilayah-wilayah terhubung dalam matriks regions menggunakan fungsi label dari modul mahotas.
# Hasilnya disimpan dalam variabel labeled, dan nr_objects berisi jumlah objek yang terlabel.

imshow(labeled, interpolation='nearest')
# Menampilkan citra labeled menggunakan fungsi imshow dari modul pylab dengan metode interpolasi 'nearest'.

show()
# Menampilkan gambar
```

## 5. Skala Interpolasi Miring 

```python
image = cv.imread("King Salman.jpg")
fig, ax = plt.subplots(1, 3, figsize=(16, 8))

# image size being 0.15 times of it's original size
image_scaled = cv.resize(image, None, fx=0.15, fy=0.15)
ax[0].imshow(cv.cvtColor(image_scaled, cv.COLOR_BGR2RGB))
ax[0].set_title("Linear Interpolation Scale")

# image size being 2 times of it's original size
image_scaled_2 = cv.resize(image, None, fx=2, fy=2, interpolation=cv.INTER_CUBIC)
ax[1].imshow(cv.cvtColor(image_scaled_2, cv.COLOR_BGR2RGB))
ax[1].set_title("Cubic Interpolation Scale")

# image size being 0.15 times of it's original size
image_scaled_3 = cv.resize(image, (200, 400), interpolation=cv.INTER_AREA)
ax[2].imshow(cv.cvtColor(image_scaled_3, cv.COLOR_BGR2RGB))
ax[2].set_title("Skewed Interpolation Scale") 
```
